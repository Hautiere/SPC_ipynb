# Statistiques

## digramme des ecarts 

Le SPC (Statistical Process Control, ou Contrôle Statistique de Processus en français) est une technique utilisée pour surveiller et contrôler la qualité d'un processus de production en utilisant des outils statistiques. Il existe plusieurs types d'analyses statistiques qui peuvent être utilisées dans le cadre du SPC, notamment :

**L'analyse des écarts** : cette analyse consiste à mesurer l'écart entre la valeur réelle d'un paramètre et sa valeur cible, afin de détecter les écarts significatifs qui pourraient indiquer un problème dans le processus.

**L'analyse de tendance** : cette analyse permet de suivre l'évolution d'un paramètre au fil du temps, afin de détecter des tendances qui pourraient indiquer un problème à venir.

**L'analyse des variations** : cette analyse consiste à mesurer la variabilité d'un paramètre dans le temps, afin de déterminer si le processus est stable ou instable.

**L'analyse des limites de contrôle** : cette analyse permet de déterminer si les variations d'un paramètre sont normales ou anormales, en utilisant des limites de contrôle définies à partir de données historiques.

Il existe également d'autres types d'analyses statistiques qui peuvent être utilisées dans le cadre du SPC, comme l'analyse de la capacité du processus ou l'analyse de la stabilité du processus.



## Visualisation des ecarts 
Voici un exemple de code Python qui peut être utilisé pour réaliser une analyse des écarts et afficher les résultats sous forme de diagramme de dispersion :

Ce code calcule d'abord les écarts entre les valeurs cibles et les valeurs réelles mesurées pour chaque paramètre, puis il utilise la fonction np.std de NumPy pour calculer l'écart-type des écarts. Le seuil d'écart est alors déterminé en multipliant cet écart-type par un facteur de 2 (ce qui correspond à un intervalle de confiance à 95 %). Enfin, le code parcours les écarts et détecte ceux qui sont supérieurs au seuil défini, en affichant un message à l'écran.

```python
import numpy as np
import matplotlib.pyplot as plt

# Valeurs cibles des paramètres
target_values = [100, 200, 300]

# Valeurs réelles mesurées pour chaque paramètre
measured_values = [99, 201, 299]

# Calcul des écarts
deltas = np.abs(np.array(measured_values) - np.array(target_values))

# Détermination des limites de contrôle à partir des données historiques
ucl, lcl = 3*np.std(deltas), -3*np.std(deltas)

# Affichage du diagramme de dispersion
plt.scatter(range(1, len(deltas)+1), deltas)
plt.hlines(ucl, xmin=1, xmax=len(deltas), linestyle='dotted')
plt.hlines(lcl, xmin=1, xmax=len(deltas), linestyle='dotted')
plt.ylabel('Écart')
plt.xlabel('Paramètre')
plt.title('Diagramme de dispersion des écarts')
plt.show()
```

## Ecart min et max

Voici un exemple de code Python qui peut être utilisé pour réaliser une analyse des écarts et afficher le résultat sous forme de graphique dans le cadre du SPC :

Ce code calcule d'abord les écarts entre les valeurs cibles et les valeurs réelles mesurées pour chaque paramètre, puis il utilise les fonctions np.min et np.max de NumPy pour déterminer les seuils d'écart minimaux et maximaux à partir des écarts calculés. Le code utilise ensuite Matplotlib pour créer un graphique représentant les écarts en fonction des paramètres, avec les seuils d'écart représentés par des lignes en pointillé. Enfin, le code affiche le graphique à l'écran.


```python
import matplotlib.pyplot as plt
import numpy as np

# Valeurs cibles des paramètres
target_values = [100, 200, 300]

# Valeurs réelles mesurées pour chaque paramètre
measured_values = [99, 201, 299]

# Calcul des écarts
deltas = np.abs(np.array(measured_values) - np.array(target_values))

# Détermination des seuils d'écart à partir des données historiques
min_threshold = 0.5*np.min(deltas)
max_threshold = 1.5*np.max(deltas)

# Création du graphique
fig, ax = plt.subplots()
ax.plot(deltas, 'o-', color='r', label='Écart')
ax.plot([0, len(deltas)-1], [min_threshold, min_threshold], '--', color='k', label='Seuil minimum')
ax.plot([0, len(deltas)-1], [max_threshold, max_threshold], '--', color='k', label='Seuil maximum')
ax.set_xlabel('Paramètre')
ax.set_ylabel('Écart')
ax.legend()
plt.show()
```
