# Boxplot & Scatter Plots analysis

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random datasets
#  loc: The mean of the normal distribution.
#  scale: The standard deviation of the normal distribution.
#  size: The number of samples
#  42 generate a random dataset with the same seed value

np.random.seed(42)
dataset1 = np.random.normal(loc=0, scale=1, size=50)
dataset2 = np.random.normal(loc=2, scale=1, size=50)

# Combine the datasets into a list
datasets = [dataset1, dataset2]

# Display the dataset
# print("Datasets:")
# print(datasets)

# Create a boxplot for each dataset
plt.boxplot(datasets)

# Set the x-axis tick labels
labels = ['Dataset 1', 'Dataset 2']
plt.xticks(range(1, len(datasets) + 1), labels)

# Set axis labels and title
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.title('Boxplot of Multiple Datasets')

# Show the boxplot
plt.show()
```

# Scatterplot

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two random datasets of size 20 for X and Y axes
np.random.seed(42)
dataset_x = np.random.normal(loc=0, scale=1, size=10)

np.random.seed(123)
dataset_y = np.random.normal(loc=0, scale=1, size=10)

# Display the dataset
print("\033[34mDataset X:\033[0m", dataset_x)
print("\033[31mDataset Y:\033[0m", dataset_y)


# Create a Scatter Plot with different colors for each dataset
plt.scatter(dataset_x, dataset_y, c='b', label='Dataset X', marker='o') # Blue color for dataset X

# We will plot the second dataset separately with the 'r' color for red
plt.scatter(dataset_y, dataset_x, c='r', label='Dataset Y', marker='s') # Red color for dataset Y

# Set axis labels and title
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('Scatter Plot')

# Show legend
plt.legend()

# Show the Scatter Plot
plt.show()
```
