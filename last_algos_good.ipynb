```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random datasets
#  loc: The mean of the normal distribution.
#  scale: The standard deviation of the normal distribution.
#  size: The number of samples
#  42 generate a random dataset with the same seed value

np.random.seed(42)
dataset1 = np.random.normal(loc=0, scale=1, size=50)
dataset2 = np.random.normal(loc=2, scale=1, size=50)

# Combine the datasets into a list
datasets = [dataset1, dataset2]

# Display the dataset
print("Datasets:")
print(datasets)

# Create a boxplot for each dataset
plt.boxplot(datasets)

# Set the x-axis tick labels
labels = ['Dataset 1', 'Dataset 2']
plt.xticks(range(1, len(datasets) + 1), labels)

# Set axis labels and title
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.title('Boxplot of Multiple Datasets')

# Show the boxplot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a random dataset
np.random.seed(42)
dataset = np.random.normal(loc=0, scale=1, size=100)

# Fit a normal distribution to the dataset
mu, std = norm.fit(dataset)

# Create a range of values for the x-axis
x = np.linspace(min(dataset), max(dataset), 100)

# Calculate the corresponding y-values for the fitted normal distribution
y = norm.pdf(x, mu, std)

# Plot the histogram of the dataset
plt.hist(dataset, bins=10, density=True, alpha=0.7)

# Plot the fitted normal distribution
plt.plot(x, y, 'r-', linewidth=2)

# Set axis labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Fitted Normal Distribution')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate two random datasets of size 10 for X and Y axes
np.random.seed(42)
dataset_x = np.random.normal(loc=0, scale=1, size=100)

np.random.seed(123)
dataset_y = np.random.normal(loc=0, scale=1, size=100)

# Display the datasets
print("\033[34mDataset X:\033[0m", dataset_x)
print("\033[31mDataset Y:\033[0m", dataset_y)

# Fit a normal distribution to the datasets
mu_x, std_x = norm.fit(dataset_x)
mu_y, std_y = norm.fit(dataset_y)

# Create a range of values for the x-axis
x = np.linspace(min(dataset_x.min(), dataset_y.min()), max(dataset_x.max(), dataset_y.max()), 100)

# Calculate the corresponding y-values for the fitted normal distributions
y_x = norm.pdf(x, mu_x, std_x)
y_y = norm.pdf(x, mu_y, std_y)

# Plot the histograms of the datasets with more transparent bars
plt.hist(dataset_x, bins=10, density=True, alpha=0.2, color='b', label='Dataset X')
plt.hist(dataset_y, bins=10, density=True, alpha=0.2, color='r', label='Dataset Y')

# Plot the fitted normal distributions
plt.plot(x, y_x, 'b-', linewidth=2, label='Fitted X')
plt.plot(x, y_y, 'r-', linewidth=2, label='Fitted Y')

# Set axis labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Fitted Normal Distributions with Transparent Bars')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a random dataset
np.random.seed(42)
dataset_x = np.random.normal(loc=0, scale=1, size=100)

# Display the dataset
print("\033[34mDataset X:\033[0m", dataset_x)

# Fit a normal distribution to the dataset
mu_x, std_x = norm.fit(dataset_x)

# Create a range of values for the x-axis
x = np.linspace(dataset_x.min(), dataset_x.max(), 100)

# Calculate the corresponding y-values for the fitted normal distribution
y_x = norm.pdf(x, mu_x, std_x)

# Calculate Upper and Lower Specification Limits (USL and LSL)
USL_x = mu_x + 3 * std_x
LSL_x = mu_x - 3 * std_x

# Plot the histogram of the dataset with more transparent bars
plt.hist(dataset_x, bins=10, density=True, alpha=0.2, color='b', label='Dataset X')

# Plot the fitted normal distribution
plt.plot(x, y_x, 'b-', linewidth=2, label='Fitted X')

# Plot vertical lines for mean, 3 sigma, USL, and LSL
plt.axvline(mu_x, color='g', linestyle='--', label='Mean')
plt.axvline(mu_x - 3 * std_x, color='orange', linestyle='--', label='Mean  3 sigma')
plt.axvline(mu_x + 3 * std_x, color='orange', linestyle='--')
plt.axvline(USL_x, color='r', linestyle='-.', label='Upper Specification Limit')
plt.axvline(LSL_x, color='r', linestyle='-.')

# Set axis labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Fitted Normal Distribution for Dataset X')

# Show legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a random dataset
np.random.seed(42)
dataset_x = np.random.normal(loc=0, scale=1, size=100)

# Display the dataset
# print("\033[34mDataset X:\033[0m", dataset_x)

# Fit a normal distribution to the dataset
mu_x, std_x = norm.fit(dataset_x)

# Create a range of values for the x-axis
x = np.linspace(dataset_x.min(), dataset_x.max(), 100)

# Calculate the corresponding y-values for the fitted normal distribution
y_x = norm.pdf(x, mu_x, std_x)

# Calculate Upper and Lower Specification Limits (USL and LSL)
USL_x = mu_x + 3 * std_x
LSL_x = mu_x - 3 * std_x

# Calculate Cp and Cpk for the dataset
Cp_x = (USL_x - LSL_x) / (6 * std_x)
Cpk_x = min((USL_x - mu_x) / (3 * std_x), (mu_x - LSL_x) / (3 * std_x))

# Plot the histogram of the dataset with more transparent bars
plt.hist(dataset_x, bins=10, density=True, alpha=0.2, color='b', label='Dataset X')

# Plot the fitted normal distribution
plt.plot(x, y_x, 'b-', linewidth=2, label='Fitted X')

# Plot vertical lines for mean, 3 sigma, USL, and LSL
plt.axvline(mu_x, color='g', linestyle='--', label='Mean')
plt.axvline(mu_x - 3 * std_x, color='orange', linestyle='--', label='Mean  3 sigma')
plt.axvline(mu_x + 3 * std_x, color='orange', linestyle='--')
plt.axvline(USL_x, color='r', linestyle='-.', label='Upper Specification Limit')
plt.axvline(LSL_x, color='r', linestyle='-.')

# Set axis labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Fitted Normal Distribution for Dataset X')

# Show legend
plt.legend()

# Show Cp and Cpk values as text in the graph
plt.text(0.05, 0.8, f'Cp = {Cp_x:.2f}', transform=plt.gca().transAxes, color='b', fontsize=12)
plt.text(0.05, 0.75, f'Cpk = {Cpk_x:.2f}', transform=plt.gca().transAxes, color='b', fontsize=12)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate a random dataset
np.random.seed(42)
dataset_x = np.random.normal(loc=0, scale=1, size=100)

# Display the dataset
print("\033[34mDataset X:\033[0m", dataset_x)

# Fit a normal distribution to the dataset
mu_x, std_x = norm.fit(dataset_x)

# Create a range of values for the x-axis
x = np.linspace(dataset_x.min(), dataset_x.max(), 100)

# Calculate the corresponding y-values for the fitted normal distribution
y_x = norm.pdf(x, mu_x, std_x)

# Calculate Upper and Lower Specification Limits (USL and LSL)
USL_x = mu_x + 3 * std_x
LSL_x = mu_x - 3 * std_x

# Calculate Cp and Cpk for the dataset
Cp_x = (USL_x - LSL_x) / (6 * std_x)
Cpk_x = min((USL_x - mu_x) / (3 * std_x), (mu_x - LSL_x) / (3 * std_x))

# Plot the histogram of the dataset with more transparent bars
n, bins, patches = plt.hist(dataset_x, bins=10, density=True, alpha=0.2, color='b', label='Dataset X')

# Plot the fitted normal distribution
plt.plot(x, y_x, 'b-', linewidth=2, label='Fitted X')

# Plot vertical lines for mean, 3 sigma, USL, and LSL
plt.axvline(mu_x, color='g', linestyle='--', label='Mean')
plt.axvline(mu_x - 3 * std_x, color='orange', linestyle='--', label='Mean  3 sigma')
plt.axvline(mu_x + 3 * std_x, color='orange', linestyle='--')
plt.axvline(USL_x, color='r', linestyle='-.', label='Upper Specification Limit')
plt.axvline(LSL_x, color='r', linestyle='-.')

# Color the histogram bars outside the specification limits in red
for i, bar in enumerate(patches):
    if bins[i] < LSL_x or bins[i+1] > USL_x:
        bar.set_facecolor('red')

# Set axis labels and title
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Fitted Normal Distribution for Dataset X')

# Show legend
plt.legend()

# Show Cp and Cpk values as text in the graph
plt.text(0.05, 0.8, f'Cp = {Cp_x:.2f}', transform=plt.gca().transAxes, color='b', fontsize=12)
```
