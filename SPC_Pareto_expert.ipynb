```python
import matplotlib.pyplot as plt

# Jeux de données
dataset_1 = [
    (4.2, 'OK'),
    (3.8, 'OK'),
    (4.5, 'KO'),
    (3.5, 'OK'),
    (4.0, 'OK'),
    (3.9, 'KO'),
    (4.1, 'OK'),
    (4.3, 'KO'),
    (3.7, 'OK'),
    (4.4, 'OK')
]

dataset_2 = [
    (7.1, 'KO'),
    (6.8, 'OK'),
    (7.0, 'KO'),
    (7.3, 'KO'),
    (7.2, 'OK'),
    (6.9, 'KO'),
    (7.4, 'OK'),
    (6.5, 'OK'),
    (7.5, 'KO'),
    (7.6, 'OK')
]

# Compter le nombre de défauts pour chaque jeu de données
def count_defects(dataset):
    return sum(1 for _, status in dataset if status == 'KO')

defects_dataset_1 = count_defects(dataset_1)
defects_dataset_2 = count_defects(dataset_2)

# Combinaison des défauts pour chaque caractéristique
defects_combined = [(defects_dataset_1, 'carac 1'), (defects_dataset_2, 'carac 2')]

# Trier les données par ordre décroissant du nombre de défauts
defects_combined.sort(reverse=True)

# Créer le graphique Pareto
defect_count = [defects for defects, _ in defects_combined]
dataset_labels = [label for _, label in defects_combined]

plt.bar(range(len(defect_count)), defect_count, align='center', alpha=0.8)
plt.xticks(range(len(defect_count)), dataset_labels)
plt.xlabel('Jeu de données')
plt.ylabel('Nombre de défauts')
plt.title('Diagramme de Pareto : Nombre de défauts par jeu de données')
plt.show()
```

# display Pareto graph with set of data

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a dataset of 40 values following a normal distribution
def generate_dataset():
    mean = np.random.uniform(3, 7)  # Random mean between 3 and 7
    std_dev = np.random.uniform(0.5, 1.5)  # Random standard deviation between 0.5 and 1.5
    return np.random.normal(mean, std_dev, 40)

# Generate 20 datasets
datasets = [generate_dataset() for _ in range(20)]

# Count the number of defects for each dataset
def count_defects(dataset):
    return sum(1 for val in dataset if val < 4 or val > 6)

defects_counts = [count_defects(dataset) for dataset in datasets]

# Sort datasets and defects_counts in descending order of defects_counts
sorted_data = sorted(zip(datasets, defects_counts), key=lambda x: x[1], reverse=True)
datasets, defects_counts = zip(*sorted_data)

# Create the Pareto chart
characteristic_labels = [f'Char {i+1}' for i in range(1, len(datasets)+1)]

plt.bar(range(len(defects_counts)), defects_counts, align='center', alpha=0.8)
plt.xticks(range(len(defects_counts)), characteristic_labels, rotation=90)
plt.xlabel('Characteristics')
plt.ylabel('Defects Count')
plt.title('Pareto Chart: Number of Defects for Each Characteristic (Descending Order)')
plt.show()
```

# display Pareto graph with synthesis line 

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate a dataset of 40 values following a normal distribution
def generate_dataset():
    mean = np.random.uniform(3, 7)  # Random mean between 3 and 7
    std_dev = np.random.uniform(0.5, 1.5)  # Random standard deviation between 0.5 and 1.5
    return np.random.normal(mean, std_dev, 40)

# Generate 20 datasets
datasets = [generate_dataset() for _ in range(20)]

# Count the number of defects for each dataset
def count_defects(dataset):
    return sum(1 for val in dataset if val < 4 or val > 6)

defects_counts = [count_defects(dataset) for dataset in datasets]

# Sort datasets and defects_counts in descending order of defects_counts
sorted_data = sorted(zip(datasets, defects_counts), key=lambda x: x[1], reverse=True)
datasets, defects_counts = zip(*sorted_data)

# Create the Pareto chart with color gradient (red to yellow to blue)
characteristic_labels = [f'Characteristic {i+1}' for i in range(1, len(datasets)+1)]
colors = plt.cm.RdYlBu(np.linspace(0, 1, len(datasets)))

plt.figure(figsize=(12, 6))
plt.bar(range(len(defects_counts)), defects_counts, align='center', alpha=0.8, color=colors)
plt.xticks(range(len(defects_counts)), characteristic_labels, rotation=90)
plt.xlabel('Characteristics')
plt.ylabel('Defects Count')
plt.title('Pareto Chart: Number of Defects for Each Characteristic (Descending Order)')

# Create the synthesis curve
total_defects = sum(defects_counts)
cumulative_percentages = np.cumsum(defects_counts) / total_defects * 100
plt.plot(range(len(defects_counts)), cumulative_percentages, marker='o', color='red', linestyle='--', label='Synthesis Curve')

plt.legend()
plt.tight_layout()
plt.show()
```
