## Parsing et affichage fichier csv
Voici comment vous pouvez extraire et afficher les données d'un fichier CSV en utilisant Python :


```python
pip install prettytable
```

```python
import csv
from prettytable import PrettyTable

# Création de la table
table = PrettyTable()

# Ouvrir le fichier CSV en mode lecture
with open('export-quimper2022.csv', 'r') as f:
  # Créer un lecteur CSV
  reader = csv.reader(f)
  
  # Ajouter les en-têtes de colonne à la table
  headers = next(reader)
  table.field_names = headers[:5]  # Prendre les 5 premiers en-têtes de colonne
  
  # Ajouter chaque ligne du fichier CSV à la table
  for row in reader:
    table.add_row(row[:5])  # Prendre les 5 premières valeurs de la ligne

# Afficher le tableau
print(table)
```
