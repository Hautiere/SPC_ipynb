## Parsing et affichage fichier csv
Voici comment vous pouvez extraire et afficher les donn√©es d'un fichier CSV en utilisant Python :


```python
pip install prettytable
```

```python
import csv
from prettytable import PrettyTable

# Cr√©ation de la table
table = PrettyTable()

# Ouvrir le fichier CSV en mode lecture
with open('export-quimper2022.csv', 'r') as f:
  # Cr√©er un lecteur CSV
  reader = csv.reader(f)
  
  # Ajouter les en-t√™tes de colonne √† la table
  headers = next(reader)
  table.field_names = headers[:5]  # Prendre les 5 premiers en-t√™tes de colonne
  
  # Ajouter chaque ligne du fichier CSV √† la table
  for row in reader:
    table.add_row(row[:5])  # Prendre les 5 premi√®res valeurs de la ligne

# Afficher le tableau
print(table)
```
